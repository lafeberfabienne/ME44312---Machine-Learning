```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('yellow_taxi_daily_trips_2009-2024.csv')
df2 = pd.read_csv('green_taxi_daily_trips_2014-2022.csv')
df3 = pd.read_csv('fhv_taxi_daily_trips_2015-2024.csv')
df4 = pd.read_csv('fhvhv_taxi_daily_trips_2019-2024.csv')



df['date'] = pd.to_datetime(df['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])
df4['date'] = pd.to_datetime(df4['date'])

```
